```python
# '수면시간'이 길수록 '수면_피로요인' 값은 작아집니다. (가장 긴 수면시간을 0으로 만듦)
final['수면피로요인'] = final['수면시간'].max() - final['수면시간']

# 2. 분석할 요인 선택 (변환된 '수면_피로요인' 사용)
# 원본 '수면시간' 컬럼은 제외합니다.
fatigue_factors = ['수면피로요인', '신체활동시간', '스트레스(1~10)', '이동거리', '이동시간']
X = final[fatigue_factors]

# 3. Z-score 표준화 적용
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
final_scaled = pd.DataFrame(X_scaled, columns=[f'{col}_scaled' for col in fatigue_factors], index=final.index)

# 4. 가중치 적용하여 '총피로도 지수' 계산
# 가중치 (수면 0.4, 스트레스 0.3, 신체활동 0.2, 이동시간 0.07, 이동거리 0.03)
weights = {
    '수면피로요인_scaled': 0.4,
    '신체활동시간_scaled': 0.2,
    '스트레스(1~10)_scaled': 0.3,
    '이동거리_scaled': 0.03,
    '이동시간_scaled': 0.07
}
final['총피로도정규화'] = sum(final_scaled[col] * weights[col] for col in final_scaled.columns)
final = pd.concat([final, final_scaled], axis=1)
final
```
